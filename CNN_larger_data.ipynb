```python
#split data into train & test for training and validation

ship_I_train_full, ship_I_test, ship_GT_train_full, ship_GT_test = train_test_split(ship_GS_Data, ship_GT, test_size=0.20, random_state=42)
back_I_train_full, back_I_test, back_GT_train_full, back_GT_test = train_test_split(back_GS_Data, back_GT, test_size=0.20, random_state=42)

X_train_full = ship_I_train_full + back_I_train_full
X_test = ship_I_test + back_I_test
y_train_full = ship_GT_train_full + back_GT_train_full
y_test = ship_GT_test + back_GT_test

X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.25, random_state=42)

np.random.seed(42)
```

```python
# network design
model = Sequential()

model.add(Conv2D(32, (3, 3), padding='same', input_shape=(80, 80, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) #40x40
model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) #20x20
model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) #10x10
model.add(Dropout(0.25))

model.add(Conv2D(32, (10, 10), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) #5x5
model.add(Dropout(0.25))

model.add(Flatten())

# fully con
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='softmax'))
```

```python
model.summary
```

```python
# optimization setup
sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)
model.compile(
    loss='binary_crossentropy',
    optimizer=sgd,
    metrics=['accuracy'])

"""# optimization setup
sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)
model.compile(
    loss='categorical_crossentropy',
    optimizer=sgd,
    metrics=['accuracy'])"""
```

```python
# training
classWeights = {0: 350, 1: 0.5}
history = model.fit(
    X_train, 
    y_train,
    epochs=5,
    validation_data=(X_val, y_val),
    verbose=2,
    class_weight = classWeights
    )
```

```python
model.evaluate(X_test, y_test)
```

```python
TP_num = 0
TP_denom = 0
TN_num = 0
TN_denom = 0
for val_index in range(len(X_test)):
    image = model.predict([X_test[val_index]])
    GT = y_test[val_index]

    if GT == 0 and image == 0:
        TN_num += 1
    if GT == 0:
        TN_denom += 1
    if GT == 1 and image == 1:
        TP_num += 1
    if GT == 1:
        TP_denom += 1
TP = TP_num / TP_denom
TN = TN_num / TN_denom
print(TP)
print(TN)
```
